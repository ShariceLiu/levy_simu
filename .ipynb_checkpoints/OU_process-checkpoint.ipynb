```python
import numpy as np
import scipy.stats as stats

dt = 0.001
ts = np.arange(0.0,5.0, dt)
N = int((5.0-0.0)/dt)
betas = np.random.normal(0.0, dt**0.5, N)
```

```python
lmda = 1
q = 1
x = 0
xs = np.zeros(N)

# mean and variance
m = x
P = 0
ms = np.zeros(N)
Ps = np.zeros(N)

for i in range(len(betas)):
    # assign
    xs[i] = x
    ms[i] = m
    Ps[i] = P
    
    # update
    dx = - lmda*x*dt + betas[i]
    x = x+dx
    m = m+(-lmda*m*dt)
    P = P+(-2*lmda*P+q)*dt

final_m = m
final_P = P
print(m, P)
```

```python
import matplotlib.pyplot as plt

plt.plot(ts, xs)
plt.plot(ts, ms+Ps**0.5)
plt.plot(ts, ms-Ps**0.5)

plt.legend(['values','mean+var','mean-var'])
```

```python
# run multiple times to check mean and variance values of the final point
lmda = 1
q = 1
x = 0

K = 1000 # number of repetitives
final_xs = np.zeros(K)

for j in range(K):
    betas = np.random.normal(0.0, dt**0.5, N)
    for i in range(len(betas)):
        # update
        dx = - lmda*x*dt + betas[i]
        x = x+dx
    
    final_xs[j]=x


# plot normal distribution
mu = final_m
var = final_P
sigma = final_P**0.5

bins = np.linspace(mu-3*sigma,mu+3*sigma, 20)
bin_width = bins[1]-bins[0]
(n, bins, patches) = plt.hist(final_xs, bins = 20)

```

```python
plt.bar(x = bins[1:]-bin_width/2, height = n/sum(n),width = bin_width )

x1s=np.linspace(mu-3*sigma,mu+3*sigma,K)
ita  = 6*sigma/bin_width
plt.plot(x1s, stats.norm.pdf(x1s,mu,sigma))
```
